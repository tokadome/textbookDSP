```python
%matplotlib inline
#%matplotlib osx

import matplotlib.pyplot as plt
import matplotlib.lines as ln
import numpy as np

import win_han
import win_ham

plt.close()

plt.rcParams['font.family'] = 'Times New Roman'

step = 1
fig = plt.figure(dpi=300)
#fig = plt.figure()
fig.patch.set_alpha(0.)

#(a)
t = np.arange(0, 5, step)
x = [-1, 2, 4, 2, -1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(221)
plt.xlim([-1, 6])
plt.ylim([-1.5, 5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.4,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$ = $h[N-n-1]$', fontsize=18, rotation=0)
plt.xticks([0])
plt.yticks([0])

markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.plot([2,2], [-1.5,5], ':', color='k')

plt.text(2.5, -2.8, '(a) Type I', ha = 'center', va = 'center', fontsize=18)

### (b)
t = np.arange(0, 6, step)
x = [-1, 2, 4, 4, 2, -1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(222)

plt.xlim([-1, 6])
plt.ylim([-1.5, 5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.4,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$ = $h[N-n-1]$', fontsize=18, rotation=0)
plt.xticks([0])
plt.yticks([0])

markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.plot([2.5, 2.5], [-1.5,5], ':', color='k')

plt.text(2.5, -2.8, '(b) Type II', ha = 'center', va = 'center', fontsize=18)
######
t = np.arange(0, 5, step)
x = [-1, 4, 0, -4, 1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(223)
plt.xlim([-1, 6])
plt.ylim([-5, 5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.44,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$ = $-h[N-n-1]$', fontsize=18, rotation=0)
plt.xticks([0])
plt.yticks([0])

markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.plot([2,2], [-5,5], ':', color='k')

plt.text(2.5, -7.0, '(c) Type III', ha = 'center', va = 'center', fontsize=18)
#####
#(d)
t = np.arange(0, 6, step)
x = [1, -2, 4, -4, 2, -1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(224)

plt.xlim([-1, 6])
plt.ylim([-5, 5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.44,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$ = $-h[N-n-1]$', fontsize=18, rotation=0)
plt.xticks([0])
plt.yticks([0])

markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.plot([2.5, 2.5], [-5,5], ':', color='k')

plt.text(2.5, -7.0, '(d) Type IV', ha = 'center', va = 'center', fontsize=18)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.6)
plt.savefig('./figures/fig10.16.eps')
plt.show()
```
