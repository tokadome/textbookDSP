```python
%matplotlib inline
#%matplotlib osx

import matplotlib.pyplot as plt
import matplotlib.lines as ln
import numpy as np

import win_han
import win_ham

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'

step = 1
fig = plt.figure(dpi=300)
#fig = plt.figure()
fig.patch.set_alpha(0.)

#(a)
t = np.arange(0, 4, step)
x = [1, 1, 1, 1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(221)
plt.xlim([-1, 6])
plt.ylim([-1.5, 1.5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.1,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$', fontsize=18, rotation=0)
plt.xticks([0,1,2,3])
plt.yticks([0,1])

markerline, stemlines, baseline = plt.stem(n,x)
#markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(2.5, -2.2, '(a)', ha = 'center', va = 'center', fontsize=18)

### (b)
t = np.arange(0, 4, step)
x = [1, 1, -1, 1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(222)

plt.xlim([-1, 6])
plt.ylim([-1.5, 1.5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.1,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$', fontsize=18, rotation=0)
plt.xticks([0,1,2,3])
plt.yticks([0,1])

markerline, stemlines, baseline = plt.stem(n,x)
#markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(2.5, -2.2, '(b)', ha = 'center', va = 'center', fontsize=18)
######
#(c)
t = np.arange(0, 5, step)
x = [1, 0.5, 0.5, -0.5, -1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(223)
plt.xlim([-1, 6])
plt.ylim([-1.5, 1.5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.1,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$', fontsize=18, rotation=0)
plt.xticks([0,1,2,3])
plt.yticks([0,1])

markerline, stemlines, baseline = plt.stem(n,x)
#markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(2.5, -2.2, '(c)', ha = 'center', va = 'center', fontsize=18)
#####
#(d)
t = np.arange(0, 5, step)
x = [1, 0.5, 0, -0.5, -1]
x = np.array(x)
n = t

ax1 =fig.add_subplot(224)

plt.xlim([-1, 6])
plt.ylim([-1.5, 1.5])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.1,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$h[n]$', fontsize=18, rotation=0)
plt.xticks([0,1,2,3])
plt.yticks([0,1])

markerline, stemlines, baseline = plt.stem(n,x)
#markerline, stemlines, baseline = plt.stem(n,x, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(2.5, -2.2, '(d)', ha = 'center', va = 'center', fontsize=18)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.6)
plt.savefig('./figures/fig9.17.png')
#plt.savefig('./figures/fig9.17.eps')
plt.show()
```
