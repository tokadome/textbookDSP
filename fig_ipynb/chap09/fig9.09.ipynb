```python
%matplotlib inline
#%matplotlib osx

import matplotlib.pyplot as plt
import matplotlib.lines as ln
import numpy as np

import win_han
import win_ham

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'

step = 1/16

t = np.arange(-5, 10, step)
x = np.cos(8*np.pi*t)+0.5*np.cos(7.2*np.pi*t)+0.25*np.cos(2*np.pi*t)

n = 16*t

fig = plt.figure(dpi=300)
fig.patch.set_alpha(0.)

#(a)
ax1 =fig.add_subplot(221)
plt.xlim([-30, 120])
plt.ylim([-2.3, 2.3])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.3,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$x[n]w_{hanning}[n]$', fontsize=18, rotation=0)
plt.yticks([-1, 0, 1])

f = win_han.extract_by_win_hanning(n, x, M=64, WB=0, fillzero=False)

markerline, stemlines, baseline = plt.stem(n,f)
#markerline, stemlines, baseline = plt.stem(n,f, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(45, -3.3, '(a)', ha = 'center', va = 'center', fontsize=18)

### (b)
ax1 =fig.add_subplot(222)
plt.xlim([-30, 120])
plt.ylim([-2.3, 2.3])
plt.axhline(0, color='k', linewidth=2)
ax1.xaxis.set_label_coords(1.04, 0.12)
ax1.yaxis.set_label_coords(0.3,1.02)
plt.xlabel('$n$', fontsize=18)
plt.ylabel('$x[n]w_{hamming}[n]$', fontsize=18, rotation=0)
plt.yticks([-1, 0, 1])

f = win_ham.extract_by_win_hamming(n, x, M=64, WB=0, fillzero=False)

markerline, stemlines, baseline = plt.stem(n,f)
#markerline, stemlines, baseline = plt.stem(n,f, color='k')
plt.setp(markerline, 'markerfacecolor', 'k')
plt.setp(stemlines, 'color', 'k')
plt.setp(baseline, 'color', 'k')

plt.text(45, -3.3, '(b)', ha = 'center', va = 'center', fontsize=18)
######
M = 64
#(c)
step = 0.001
omg = np.arange(-2*np.pi, 2*np.pi, step)

omg1 = np.pi/2
omg2 = 1.9*np.pi/4
omg3 = np.pi/8

ax2 =fig.add_subplot(223)
X1 = 0.5*win_han.win_hanning_spectral(64,omg1,-2*np.pi,2*np.pi,step) + 0.5*win_han.win_hanning_spectral(64,-omg1,-2*np.pi,2*np.pi,step)
X2 = 0.5*win_han.win_hanning_spectral(64,omg2,-2*np.pi,2*np.pi,step) + 0.5*win_han.win_hanning_spectral(64,-omg2,-2*np.pi,2*np.pi,step)
X3 = 0.5*win_han.win_hanning_spectral(64,omg3,-2*np.pi,2*np.pi,step) + 0.5*win_han.win_hanning_spectral(64,-omg3,-2*np.pi,2*np.pi,step)

X = X1 + 0.5*X2 + 0.25*X3

X_gain = np.absolute(X)
X_gain_max = max(X_gain)

ax2 =fig.add_subplot(223)
plt.xlim([-np.pi, np.pi])
plt.ylim([-0.1, 1.2])
plt.axhline(0, color='k', linewidth=1)

ax2.xaxis.set_label_coords(1.05, 0.1)
ax2.yaxis.set_label_coords(0.27,1.02)
plt.ylabel('$|X(\omega)|/|X_{max}|$', fontsize=18, rotation=0)
plt.xlabel('$\omega$', fontsize=18, rotation=0)
plt.xticks([-3,-2,-1, 0,1,2, 3])
plt.yticks(range(0),())

plt.plot(omg, X_gain/X_gain_max, color='k')

plt.text(0, -0.4, '(c)', ha = 'center', va = 'center', fontsize=18)

#####
#(d)
ax2 =fig.add_subplot(224)

X1 = 0.5*win_ham.win_hamming_spectral(64,omg1,-2*np.pi,2*np.pi,step) + 0.5*win_ham.win_hamming_spectral(64,-omg1,-2*np.pi,2*np.pi,step)
X2 = 0.5*win_ham.win_hamming_spectral(64,omg2,-2*np.pi,2*np.pi,step) + 0.5*win_ham.win_hamming_spectral(64,-omg2,-2*np.pi,2*np.pi,step)
X3 = 0.5*win_ham.win_hamming_spectral(64,omg3,-2*np.pi,2*np.pi,step) + 0.5*win_ham.win_hamming_spectral(64,-omg3,-2*np.pi,2*np.pi,step)

X = X1 + 0.5*X2 + 0.25*X3

X_gain = np.absolute(X)
X_gain_max = max(X_gain)

ax2 =fig.add_subplot(224)
plt.xlim([-np.pi, np.pi])
plt.ylim([-0.1, 1.2])
plt.axhline(0, color='k', linewidth=1)
#ax2 = plt.gca()
ax2.xaxis.set_label_coords(1.05, 0.1)
ax2.yaxis.set_label_coords(0.27,1.02)
plt.ylabel('$|X(\omega)|/|X_{max}|$', fontsize=18, rotation=0)
plt.xlabel('$\omega$', fontsize=18, rotation=0)
plt.xticks([-3,-2,-1, 0,1,2, 3])
plt.yticks(range(0),())

plt.plot(omg, X_gain/X_gain_max, color='k')

plt.text(0, -0.4, '(d)', ha = 'center', va = 'center', fontsize=18)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.6)
plt.savefig('./figures/fig9.09.png')
#plt.savefig('./figures/fig9.09.eps')
plt.show()
```
